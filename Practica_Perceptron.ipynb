```python
import numpy as np

# Inicializar pesos y umbral aleatorios
W = np.random.uniform(-1, 1, 2)  # Dos entradas
theta = np.random.uniform(-1, 1)
alfa = 0.1  # Tasa de aprendizaje
error_aceptable = 0.01  # Umbral de error

# Generar un dataset de 100 puntos (x1, x2) en el rango [0, 1]
np.random.seed(0)
n_samples = 100
X_data = np.random.rand(n_samples, 2)
D_data = np.array([1 if x[1] > 0.5 * x[0] + 0.2 else 0 for x in X_data])
dataset = [(X_data[i], D_data[i]) for i in range(n_samples)]

# Funciones de activación y cálculo de error
def funcion_activacion(valor):
    return 1 if valor >= 0 else 0

def calcular_salida(X, W, theta):
    y_in = np.dot(X, W) - theta
    Y = funcion_activacion(y_in)
    return Y

def calcular_error(deseado, Y):
    return deseado - Y

def actualizar_pesos(W, X, error, alfa):
    return W + alfa * error * X

# Bucle de entrenamiento
for epoch in range(100):  # Número máximo de iteraciones
    total_error = 0
    for X, deseado in dataset:
        Y = calcular_salida(X, W, theta)
        error = calcular_error(deseado, Y)
        W = actualizar_pesos(W, X, error, alfa)
        total_error += abs(error)
    if total_error < error_aceptable:
        print("Entrenamiento completado")
        break
    print(f"Época {epoch+1}, Error total: {total_error}")

print("Pesos finales:", W)
print("Umbral final:", theta)
```

```python
import matplotlib.pyplot as plt
x1_vals = np.linspace(0, 1, 100)
x2_vals = -(W[0] * x1_vals - theta) / W[1]  # Ecuación de la frontera de decisión


plt.figure(figsize=(8, 6))
for X, D in dataset:
    if D == 0:
        plt.scatter(X[0], X[1], color="blue", label="Clase 0" if D == 0 else "", marker="o")
    else:
        plt.scatter(X[0], X[1], color="red", label="Clase 1" if D == 1 else "", marker="x")

# Graficar la línea de frontera de decisión
plt.plot(x1_vals, x2_vals, color="green", label="Frontera de decisión")

# Configurar etiquetas y leyenda
plt.xlabel("x1")
plt.ylabel("x2")
plt.title("Dataset y frontera de decisión aprendida")
plt.legend(["Frontera de decisión", "Clase 0", "Clase 1"])
plt.grid(True)
plt.show()
```
